Dans Ce Projet nous souhaitons mettre en place un système de détection de la fatigue grâce au calcul de "eye aspect ratio" (EAR). L'idée est de suivre les valeurs de l'EAR au fil du temps et de définir un seuil minimum pour détecter la durée des yeux fermés (indicateur de la fatigue).
L’image suivante montre deux états: l'oeil ouvert et fermé avec les le positionnement des points d'intérêts (p1,p2,...,p6):

<div style="text-align:center; margin-top: 20px;">
    <img src="assets/eye1.png" alt="Eye 1"style="margin-right: 20px;">
    <img src="assets/eye2.png" alt="Eye 2">
</div>

l'expression de l'EAR est donnée par la formule suivante:

$$\text{EAR} = \frac{\|p2 - p6\| + \|p3 - p5\|}{2\|p1 - p4\|} $$




pour réaliser ce projet, nous allons suivre les étapes suivantes:

1. Extraction des images à partir d'une vidéo.
2. Conversion des images en niveaux de gris.
3. Suppression de bruits.
4. Amélioration de la qualité des images (contraste, luminosité).
5. Détection des contours (utilisant des méthodes telles que Canny ou Sobel).
6. Détection des yeux.
7. Application de techniques de morphologie.
8. Récupération des points d'intérêts (p1,...p6).
9. À chaque instant t (par exemple, t=10s), calcul de l'EAR et sauvegarde dans un fichier CSV.
10. Détection de la fatigue en utilisant les valeurs de l'EAR.

tous d'abord, nous allons commencer par importer les bibliothèques nécessaires pour réaliser ce projet.

```python
import csv
import cv2
import dlib
from scipy.spatial import distance
import pygame
import time
import os
import numpy as np

import matplotlib.pyplot as plt
```

En raison de simplicité j'ai esseillé de divisé le code en plusieurs fonctions pour faciliter la lecture et la compréhension.
la première fonction est la fonction qui permet de calculer ***l'EAR***.
La fonction prend en paramètre une liste des coordonnées des points d'intérêts et retourne la valeur de l'EAR.
le calcule de la norme suivant la formule donnée ci-dessus.
$$\text{EAR} = \frac{\|p2 - p6\| + \|p3 - p5\|}{2\|p1 - p4\|} $$

Pour calculer la norme on utilise la fonction 
$$\|\mathbf{a} - \mathbf{b}\| = \sqrt{(a_1 - b_1)^2 + (a_2 - b_2)^2}$$
qui est la fonction
***distance.euclidean*** de la bibliothèque ***scipy.spatial***.



```python
def eye_aspect_ratio(eye):
    # Compute the euclidean distances between the two sets of vertical eye landmarks (x, y)-coordinates
    A = distance.euclidean(eye[1], eye[5])
    B = distance.euclidean(eye[2], eye[4])
    # Compute the euclidean distance between the horizontal eye landmark (x, y)-coordinates
    C = distance.euclidean(eye[0], eye[3])
    # Compute the eye aspect ratio
    ear = (A + B) / (2.0 * C)
    # Return the eye aspect ratio
    return ear
```

Maintenant et pour continuer avec la première partie du projet, nous allons commencer par la détection des yeux et le visage , et cela ne peut pas etre faite sans extraire les frames de la vidéo.
c'est pour cela que nous allons creer une fonction qui permet de **Capturer toutes les frames de la video** et aussi de **construire une liste des frames choisie pour chaque intervale de temps donné.**

```python
def split_video_frames(video_path):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Get the frames per second (fps) and frame width/height
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Calculate the frame interval for every 2 seconds
    frame_interval = int(fps * 2)

    # Lists to store selected frames and all frames
    selected_frames = []
    all_frames = []

    # Counter for frame number
    frame_count = 0

    while True:
        # Read a frame from the video
        ret, frame = cap.read()

        # Break the loop if we have reached the end of the video
        if not ret:
            break

        # Append the frame to the all_frames list
        all_frames.append(frame)

        # If the frame count is a multiple of frame_interval, add it to the selected frames
        if frame_count % frame_interval == 0:
            selected_frames.append(frame)

        # Increment the frame count
        frame_count += 1

    # Release the video capture object
    cap.release()

    # Return the selected frames, all frames, and frame rate
    return selected_frames, all_frames, fps, frame_width, frame_height
```

Étant donné que cette fonction génère une liste de toutes les images de la vidéo et une autre liste contenant les images sélectionnées à un intervalle de temps spécifique, il est maintenant opportun de développer deux fonctions distinctes. La première fonction sera dédiée au **prétraitement des images sélectionnées** et à la détection de la fatigue dans ces images. La seconde fonction sera chargée de la **détection des yeux et du visage tout au long de la vidéo**.

La première fonction est la fonction ***image_processing*** qui prend en paramètre la liste des images sélectionnées et effectue les opérations suivantes sur chaque image:
pour garder les choses strictures, chaque prétraitement sera stocké dans un dossier spécifique.

```python
def image_processing(selected_frames):
    # Create a new directory to store the processed frames
    os.makedirs("output_frames", exist_ok=True)
    os.makedirs("output_frames/canny", exist_ok=True)
    os.makedirs("output_frames/gray", exist_ok=True)
    os.makedirs("output_frames/morphologie", exist_ok=True)
    os.makedirs("output_frames/bilaterale", exist_ok=True)
    os.makedirs("output_frames/egalization", exist_ok=True)
    os.makedirs("output_frames/luminosite", exist_ok=True)
    print("folders created")
    # Process each selected frame
    for i, frame in enumerate(selected_frames):
        # Convert the frame to grayscale
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        bilateral = cv2.bilateralFilter(gray, 9, 75, 75)
        canny = cv2.Canny(gray, 50, 150)
        morph = cv2.morphologyEx(canny, cv2.MORPH_CLOSE, (5, 5))
        # améliorer le contraste global de l'image.
        egalisation = cv2.equalizeHist(gray)
        # améliorer la luminosité de l'image.
        luminosite = cv2.convertScaleAbs(gray, alpha=1.2, beta=30)

        # Save the processed frame to the new directory
        cv2.imwrite(f"output_frames/gray/frame_{i}.jpg", gray)
        cv2.imwrite(f"output_frames/bilaterale/frame_{i}.jpg", bilateral)
        cv2.imwrite(f"output_frames/canny/frame_{i}.jpg", canny)
        cv2.imwrite(f"output_frames/morphologie/frame_{i}.jpg", morph)
        cv2.imwrite(f"output_frames/egalization/frame_{i}.jpg", egalisation)
        cv2.imwrite(f"output_frames/luminosite/frame_{i}.jpg", luminosite)
```

Dans le processus de détection de la fatigue, nous allons enregistrer chaque valeur de l'EAR dans un fichier CSV pour pouvoir l'utiliser ultérieurement.

l'utilité de cette fonction est de sauvegarder les valeurs de l'EAR dans un fichier CSV, qu'il sera possible de visualiser et d'analyser ultérieurement.

![analyse](assets/analytics.png)

La fonction ***save_ear_to_csv*** prend en paramètre une liste de valeurs de l'EAR et enregistre ces valeurs dans un fichier CSV.
Pour que Nous puissions visualiser les valeurs de l'EAR détectées dans chaque frame et aussi ceux détecter comme étant des frames de fatigue.

cette fonction enregistre les valeurs de l'EARS et aussi la date et l'heure de chaque valeur dans un fichier CSV.

```python
def save_to_csv(ear, filename):
    current_time = time.time()
    # Convertir le temps actuel en une structure de temps locale
    local_time = time.localtime(current_time)

    # Extraire les composants de l'heure
    hours = local_time.tm_hour
    minutes = local_time.tm_min
    seconds = local_time.tm_sec

    # Save the ear values to a CSV file
    with open(filename, 'a', newline='') as file:
        writer = csv.writer(file)
        if file.tell() == 0:
            writer.writerow(['Ear', 'Time'])
        writer.writerow([ear, f"{int(hours)}:{int(minutes)}:{int(seconds)}"])
    file.close()
```

Alors que maintenant On'est pret pour la deuxième fonction qui est la fonction ***Video_detection*** qui prend en paramètre la localisation de la vidéo puis elle appelle les fonctions précédentes pour effectuer le prétraitement des images sélectionnées et la détection de la fatigue.

Cette fonction dans son travail necessite l'utilisation de la bibliothèque ***dlib*** pour la détection des visages et des yeux.
Cette bibliothèque offre des fonctions des modèles pré-entrainés pour la détection des visages et donner des points d'intérêts dans le visage.
Alors il nous faut premièrement initialiser le détecteur de visage et le détecteur de points d'intérêts.

```python
# initialize the Model de detection de visage 
face_detection = dlib.get_frontal_face_detector()

# initialize the Model de detection des points d'intérêts
face_landmark = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
```

les Landmarks sont des points d'intérêts dans le visage, ils sont utilisés pour détecter les yeux, le nez, la bouche, etc. il ont numéroté de 0 à 67 de la façon suivante:

<div style="text-align:center; margin-top: 20px;">
    <img src="assets/OIP.jpg" alt="Landmarks">

```python
def Video_detection(video_path):
    # take a video and split it into frames:
    selected_frames, all_frames, frame_rate, frame_width, frame_height = split_video_frames(video_path)

    # Generate a new video with face and age detection
    # first create a new directory to store the output video and also initialize the video writer
    os.makedirs("output_videos", exist_ok=True)
    output_video_path = f"output_videos/video{time.time()}.mp4"
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video_path, fourcc, frame_rate, (frame_width, frame_height))
    
    #initialize the counter : when the counter is greater than 3, we will consider that the person is fatigued
    counter = []
    
    # Process each selected frame : calling the image_processing function
    image_processing(selected_frames)
    
    # Process each frame in the video
    
    for frame in all_frames:
        # Convert the frame to grayscale
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        # detect faces in the frame:
        faces = face_detection(gray)
        # Process each detected face
        for face in faces:
            face_landmarks = face_landmark(gray, face)
            eye = []
            # Extract the eye coordinates for only eyes
            for n in range(36, 48):
                x = face_landmarks.part(n).x
                y = face_landmarks.part(n).y
                # Draw a circle around the eye
                cv2.circle(frame, (x, y), 2, (0, 255, 0), 1)
                # Append the eye coordinates to the eye list : the first 6 points are the points of interest for the first eye and the second 6 points are the points of interest for the second eye in forme of (x,y): those cordinates are used to calculate the EAR
               
                eye.append((x, y))
            # Draw a rectangle around the detected face
            x1 = face.left()
            y1 = face.top()
            x2 = face.right()
            y2 = face.bottom()
            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)

            # Calculate the eye aspect ratio for the detected face
            ear = eye_aspect_ratio(eye)
            # Save all ear values to a CSV file
            os.makedirs("output_ear_data", exist_ok=True)
            save_to_csv(ear, 'output_ear_data/ear.csv')
            
            # Detect fatigue using the eye aspect ratio and a treshold
            if ear < 0.285:
                # number of frames the person is detected as fatigued
                counter.append(1)
                if len(counter) > 3:
                    # Save the detected fatigue data to a CSV file
                    save_to_csv(ear, 'output_ear_data/detected_fatigue_data.csv')
                    # Draw a text label for the detected fatigue as an alert
                    cv2.putText(frame, "Fatigue Detected", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
            else:
                counter = []
            print(ear, "\t", counter)
        # Save the frame to the output video
        out.write(frame)

    out.release()
```

# Part 2 : Détection de la fatigue en temps réel en utilisant la webcam

Dans cette partie, nous allons développer une application pour détecter la fatigue **en temps réel** en utilisant la webcam. Nous allons utiliser la bibliothèque ***OpenCV*** pour capturer des images à partir de la webcam et la bibliothèque ***dlib*** pour détecter les visages et les yeux dans les images capturées.

Cette application fonctionnera de manière similaire à la détection de la fatigue à partir d'une vidéo, mais au lieu de traiter des images à partir d'une vidéo, elle traitera des images capturées en temps réel à partir de la webcam. Elle est basée sur le même logique que la première partie du projet et la meme fonction de calcul de l'EAR.

Puisque les models de détection de visage et de points d'intérêts sont déjà initialisés, il nous faut juste initialiser la webcam et capturer des images à partir de celle-ci.

```python
cap = cv2.VideoCapture(0)
# re initialiser le counter
counter = []
```

Pour cette partie, chaque fois que nous capturons une image à partir de la webcam, nous lanceons un son d'alerte pour avertir la personne de la fatigue.
cette fonction est réalisée en utilisant la bibliothèque ***pygame***.

```python
# Initialize pygame
pygame.init()

# Load the sound file
sound_file = "alert.wav"
pygame.mixer.music.load(sound_file)

# Set the volume (optional)
pygame.mixer.music.set_volume(1.0)
```

Et alors la fonction qui declanche une fenetre de camera pour la detection au temps réel de la fatigue.

```python
while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_detection(gray)
    # bouclé sur les visages détectés
    for face in faces:
        face_landmarks = face_landmark(gray, face)
        eye = []
        
        for n in range(36, 48):
            x = face_landmarks.part(n).x
            y = face_landmarks.part(n).y
            cv2.circle(frame, (x, y), 2, (0, 255, 0), 1)
            eye.append((x, y))
        # cette partie est pour dessiner les yeux avec des lignes enchainées
        # Convert the list to a NumPy array
        eye_array = np.array(eye, dtype=np.int32)

        # Reshape the array to have two sets of 6 coordinates (2 eyes, each with 6 landmarks)
        eye_array = eye_array.reshape(2, 6, 2)
        for eye_landmarks in eye_array:
            cv2.polylines(frame, [eye_landmarks], isClosed=True, color=(0, 255, 255), thickness=1)

        # Draw rectangles around the face
        x1 = face.left()
        y1 = face.top()
        x2 = face.right()
        y2 = face.bottom()
        cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
        ear = eye_aspect_ratio(eye)

        os.makedirs("output_ear_data", exist_ok=True)
        save_to_csv(ear, 'output_ear_data/ear_live.csv')
        if ear < 0.24:
            counter.append(1)
            if len(counter) > 2:
                # Save the detected fatigue data to a CSV file
                save_to_csv(ear, 'output_ear_data/detected_fatigue_data_live.csv')
                pygame.mixer.music.play()
                pygame.time.wait(200)
                cv2.putText(frame, "Fatigue Detected", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
        else:
            counter = []
        print(counter)

    cv2.imshow("Frame", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```
